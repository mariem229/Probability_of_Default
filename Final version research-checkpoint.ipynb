### Logistic Regression

```python
loan.columns.tolist
```

```python
loan.sample(5)
```

```python
cols = [col for col in loan.columns if col != 'loan_status']
cols.append('loan_status')
loan = loan[cols]
loan.sample(5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

accuracies = []
y_real = []
y_proba = []
y_pred_all = []

kf = KFold(n_splits=10, shuffle=True, random_state=42)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Training the logistic regression model
    log_reg = LogisticRegression(penalty='l2', C=5.0, solver='liblinear')
    log_reg.fit(X_train, y_train)
    
    # Predicting labels
    y_pred = log_reg.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))
    y_pred_all.extend(y_pred)

    # Predicting probabilities for ROC Curve
    y_pred_proba = log_reg.predict_proba(X_test)[:, 1]
    y_real.extend(y_test)
    y_proba.extend(y_pred_proba)

# Calculate Mean Accuracy
mean_accuracy = np.mean(accuracies)
print(f"Mean Accuracy: {mean_accuracy}")

# Calculate ROC Curve and AUC
fpr, tpr, _ = roc_curve(y_real, y_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - Logistic Regression')
plt.legend(loc="lower right")
plt.show()

# Calculate Confusion Matrix
cm = confusion_matrix(y_real, y_pred_all)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()

```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=10, max_depth=5, max_features='sqrt', bootstrap=True)
```

```python
accuracies = []
y_real = []
y_proba = []

kf = KFold(n_splits=10, shuffle=True, random_state=42)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))

    # For ROC Curve
    proba = model.predict_proba(X_test)[:, 1]
    y_real.extend(y_test)
    y_proba.extend(proba)

# Mean Accuracy
mean_accuracy = np.mean(accuracies)
print(f"Mean Accuracy: {mean_accuracy}")

# Calculate ROC Curve and AUC
fpr, tpr, _ = roc_curve(y_real, y_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - Random Forest')
plt.legend(loc="lower right")
plt.show()

# Confusion Matrix
y_pred_all = [1 if x > 0.5 else 0 for x in y_proba]
cm = confusion_matrix(y_real, y_pred_all)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix- Random Forest')
plt.show()
```

### XG boost

```python
import xgboost as xgb
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
X = loan.drop('loan_status', axis=1)
y = loan['loan_status']

accuracies = []
y_real = []
y_proba = []
y_pred_all = []

kf = KFold(n_splits=10, shuffle=False)

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Initialize XGBoost Classifier
    xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

    # Fit the model
    xgb_model.fit(X_train, y_train)

    # Predictions for accuracy
    y_pred = xgb_model.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))
    y_pred_all.extend(y_pred)

    # Predictions for ROC Curve
    y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]
    y_real.extend(y_test)
    y_proba.extend(y_pred_proba)

# Calculate Mean Accuracy
mean_accuracy = np.mean(accuracies)
print(f"Mean Accuracy: {mean_accuracy}")

# Calculate ROC Curve and AUC
fpr, tpr, _ = roc_curve(y_real, y_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - XGBoost')
plt.legend(loc="lower right")
plt.show()

# Calculate Confusion Matrix
cm = confusion_matrix(y_real, y_pred_all)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - XGBoost')
plt.show()
```
